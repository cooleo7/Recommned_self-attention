```python
mh = pd.read_csv('mh_img1.csv')
mh
```

```python
a_df = mh.query('model =="SASRec"') 
#a_visits = a_df['Precision 1']
#days = a_df['days']
 
b_df = mh.query('model =="TiSASRec"')
#b_visits = b_df['visits']
 
c_df = mh.query('model =="BERT4Rec"')
d_df = mh.query('model =="MEANTIME"')
e_df = mh.query('model =="CSE"')
```

```python
#import matplotlib.pyplot as plt
fig = pltpy.figure(figsize=(12, 8)) #    ( 14,  5)
fig.suptitle('Performance comparison by the number of head')

ax1 = fig.add_subplot(2, 3, 1) 
ax2 = fig.add_subplot(2, 3, 2) 
ax3 = fig.add_subplot(2, 3, 3) 
ax4 = fig.add_subplot(2, 3, 4) 
ax5 = fig.add_subplot(2, 3, 5) 

# [0, 1]   
ax1.plot(a_df['head'],a_df['Precision 1'], marker = 'o', markersize = 5, label = 'Precision 1')
ax1.plot(a_df['head'],a_df['Precision 10'], marker = 'o', markersize = 5, label = 'Precision 10')
ax1.plot(a_df['head'],a_df['Recall 1'], marker = 'o', markersize = 5, label = 'Recall 1')
ax1.plot(a_df['head'],a_df['Recall 10'], marker = 'o', markersize = 5, label = 'Recall 10')
ax1.plot(a_df['head'],a_df['MRR 20'], marker = 'o', markersize = 5, label = 'MRR 20')
ax1.plot(a_df['head'],a_df['NDCG 20'], marker = 'o', markersize = 5, label = 'NDCG 20')
ax1.set_title('SARec', fontsize = 12)
#ax1.legend()

# [1, 3]   
ax2.plot(b_df['head'],b_df['Precision 1'], marker = 'o', markersize = 5, label = 'Precision 1')
ax2.plot(b_df['head'],b_df['Precision 10'], marker = 'o', markersize = 5, label = 'Precision 10')
ax2.plot(b_df['head'],b_df['Recall 1'], marker = 'o', markersize = 5, label = 'Recall 1')
ax2.plot(b_df['head'],b_df['Recall 10'], marker = 'o', markersize = 5, label = 'Recall 10')
ax2.plot(b_df['head'],b_df['MRR 20'], marker = 'o', markersize = 5, label = 'MRR 20')
ax2.plot(b_df['head'],b_df['NDCG 20'], marker = 'o', markersize = 5, label = 'NDCG 20')
ax2.set_title('TiSASRec', fontsize = 12)
#ax2.legend()

ax3.plot(c_df['head'],c_df['Precision 1'], marker = 'o', markersize = 5, label = 'Precision 1')
ax3.plot(c_df['head'],c_df['Precision 10'], marker = 'o', markersize = 5, label = 'Precision 10')
ax3.plot(c_df['head'],c_df['Recall 1'], marker = 'o', markersize = 5, label = 'Recall 1')
ax3.plot(c_df['head'],c_df['Recall 10'], marker = 'o', markersize = 5, label = 'Recall 10')
ax3.plot(c_df['head'],c_df['MRR 20'], marker = 'o', markersize = 5, label = 'MRR 20')
ax3.plot(c_df['head'],c_df['NDCG 20'], marker = 'o', markersize = 5, label = 'NDCG 20')
ax3.set_title('BERT4Rec', fontsize = 12)
#ax3.legend(loc='best')

ax4.plot(d_df['head'],d_df['Precision 1'], marker = 'o', markersize = 5, label = 'Precision 1')
ax4.plot(d_df['head'],d_df['Precision 10'], marker = 'o', markersize = 5, label = 'Precision 10')
ax4.plot(d_df['head'],d_df['Recall 1'], marker = 'o', markersize = 5, label = 'Recall 1')
ax4.plot(d_df['head'],d_df['Recall 10'], marker = 'o', markersize = 5, label = 'Recall 10')
ax4.plot(d_df['head'],d_df['MRR 20'], marker = 'o', markersize = 5, label = 'MRR 20')
ax4.plot(d_df['head'],d_df['NDCG 20'], marker = 'o', markersize = 5, label = 'NDCG 20')
ax4.set_title('MEANTIME', fontsize = 12)
#ax4.legend(loc='best')

ax5.plot(e_df['head'],e_df['Precision 1'], marker = 'o', markersize = 5, label = 'Precision 1')
ax5.plot(e_df['head'],e_df['Precision 10'], marker = 'o', markersize = 5, label = 'Precision 10')
ax5.plot(e_df['head'],e_df['Recall 1'], marker = 'o', markersize = 5, label = 'Recall 1')
ax5.plot(e_df['head'],e_df['Recall 10'], marker = 'o', markersize = 5, label = 'Recall 10')
ax5.plot(e_df['head'],e_df['MRR 20'], marker = 'o', markersize = 5, label = 'MRR 20')
ax5.plot(e_df['head'],e_df['NDCG 20'], marker = 'o', markersize = 5, label = 'NDCG 20')
ax5.set_title('CSE', fontsize = 12)
ax5.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

pltpy.savefig('./final_mh.png')
pltpy.show()
```
